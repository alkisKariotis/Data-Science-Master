```python
# Libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

#Import some libraries to make displaying easier
import warnings
warnings.filterwarnings('ignore')

pd.get_option("display.max_columns")
```

## a. (10 pts) Load and use the above data set, but do not use the class labels

```python
# Read the CSV file from the URL
df = pd.read_csv('wine-quality.csv', sep=";")

# Drop rows with missing values
df = df.dropna()

# Drop duplicates
df = df.drop_duplicates()

# Split the data into input and target / Discard the class labels
X = df.drop("quality", axis=1)

X
```

## b. (10 pts) Apply the k-Means clustering algorithm; experiment with a different number of cluster
centres and discover the best inertia and silhouette values that you can get.

c. (10 pts) Print the clustering results with your code: number of data items per cluster, quality of
clustering measured by the inertia and silhouette values for the best parameters for k-means

```python
# Experiment with different numbers of clusters to find the best inertia and silhouette scores
best_inertia = None
best_silhouette = None
best_clusters = None

for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    inertia = kmeans.inertia_
    silhouette = silhouette_score(X, kmeans.labels_)
    cluster_counts2 = pd.Series(best_clusters).value_counts()
    print(f"Number of data items per cluster: {cluster_counts2.to_dict()}")
    print(f"k={n_clusters}, inertia={inertia:.2f}, silhouette={silhouette:.2f}")
    if best_inertia is None or inertia < best_inertia:
        best_inertia = inertia
        best_silhouette = silhouette
        best_clusters = kmeans.labels_

# Print the best clustering results
cluster_counts = pd.Series(best_clusters).value_counts()
print(f"Best clustering with k={len(cluster_counts)}, inertia={best_inertia:.2f}, silhouette={best_silhouette:.2f}")
print(f"Number of data items per cluster: {cluster_counts.to_dict()}")
```
