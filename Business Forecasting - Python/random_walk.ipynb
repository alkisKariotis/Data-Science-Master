The random function generates a uniform distribution of dumbers between 0 and 1. The mean of these numbers is (0+1)/2=0.5. If i multiply the series by 2 and substract 1 then the mean of the series is 0 2*0.5-1

```python
series = 2*np.random.random((1000,1)) - 1
series.mean()
```

```python
r_walk=np.cumsum(series)
```

```python
random_walk=pd.DataFrame(r_walk)
```

```python
random_walk.rename(columns = {0:'RW'}, inplace = True)
```

```python
random_walk.plot()
```

```python
plot_acf(random_walk)
```

```python
plot_pacf(random_walk)
```
