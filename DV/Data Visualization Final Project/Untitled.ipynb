```python
import pandas as pd
import numpy as np

# Load the dataset
airbnb_df = pd.read_csv("AB_NYC_2019.csv")

# Select the columns we are interested in
price = airbnb_df['price']
min_nights = airbnb_df['minimum_nights']

# Calculate the correlation coefficient
corr_coef = np.corrcoef(price, min_nights)[0][1]

print("The correlation coefficient between price and minimum nights is:", corr_coef)
```

```python
import pandas as pd
import numpy as np

# Load the dataset
airbnb_df = pd.read_csv("AB_NYC_2019.csv")

# Select the columns we are interested in
price = airbnb_df['price']
min_nights = airbnb_df['minimum_nights']

# Calculate the predicted prices using the y_fit formula
y_fit = 143.76 + 0.64 * min_nights

# Calculate the mean squared error
mse = np.mean((price - y_fit)**2)

print("The mean squared error between the actual prices and the predicted prices is:", mse)
```

```python
import pandas as pd
import numpy as np

# Create DataFrame with 48895 rows
rating = pd.DataFrame({'rating': np.random.randint(1, 6, size=48895)})

# Load the dataset
airbnb_df = pd.read_csv("AB_NYC_2019.csv")

# Merge the two DataFrames
merged_df = airbnb_df.join(rating)

# Sort the merged DataFrame by rating in descending order and then by number_of_reviews
sorted_df = merged_df.sort_values(['rating', 'number_of_reviews'], ascending=[False, False])

# Filter the DataFrame to include only rows where 'number_of_reviews' is not 0
filtered_df = sorted_df[sorted_df['number_of_reviews'] != 0]

# Print the filtered DataFrame
filtered_df.head(5)
```
