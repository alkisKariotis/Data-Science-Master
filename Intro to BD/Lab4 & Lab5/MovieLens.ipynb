## Read the u.item FIle

```python
# Define the column names
columns = ["movie id", "movie title", "release date", "video release date",
           "IMDb URL", "unknown", "Action", "Adventure", "Animation",
           "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy",
           "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi",
           "Thriller", "War", "Western"]

# Read the CSV file with double quotes as the delimiter
df_item = pd.read_csv('u.item', sep='|',index_col=None, header=None, names=columns, quoting=3,encoding='latin1')
df_item
```

## Read the u.data File

```python
# Read the CSV of users
columns_user= ['user id','item id','rating','timestamp'] 
df_data = pd.read_csv('u.data', sep='\t',index_col=None, header=None, names=columns_user)
df_data
```

```python
df_user = df_user.drop(['age', 'gender','zip code'], axis=1)
df_user.columns
```

```python
df_data = df_data.drop(['rating', 'timestamp'], axis=1)
df_data.columns
```

```python
merged_user_data = pd.merge(df_user, df_data, on='user id', how = 'inner')
merged_user_data = merged_user_data.rename(columns={'item id': 'movie id'})
merged_user_data
```

```python
final_merge = pd.merge(merged_user_data, df_item, on='movie id', how = 'inner')
final_merge
```

```python
final_merge.groupby('occupation')['movie id'].count().reset_index()
```
